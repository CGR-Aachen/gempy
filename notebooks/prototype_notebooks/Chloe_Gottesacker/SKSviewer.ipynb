### Viewer for checking SKS inputs and outputs

```python
import numpy as np
import pandas as pn
import matplotlib.pyplot as plt
import matplotlib

#embed matplotlib figures into notebooks:
%matplotlib inline     
```

```python
#Set path to SKS outputs:
path = r'C:\Users\Chloe\Desktop\PhD\SKS\output_SKS'
node_filename =         path + '\output_SKS_simpleNodes.txt'
pipe_filename =         path + '\output_SKS_simpleLinks.txt'
complex_node_filename = path + '\output_SKS_nodes'
```

```python
#Read in complex network node data:
nodes = pn.read_csv(complex_node_filename, sep=' ', names=['X','Y','Z'])
print(nodes.head())
plt.scatter(nodes.X,nodes.Y, marker='.')
```

```python
#Read in simple node coordinate data
nodes = pn.read_csv(node_filename, delim_whitespace=True, names=['X','Y'])
print(nodes.head())
plt.scatter(nodes.X,nodes.Y, marker='.')
for i in range(len(nodes)):
    plt.annotate(nodes.index[i],xy=(nodes.X[i],nodes.Y[i]))
```

```python
#Read in segment connection data:
pipes = pn.read_csv(pipe_filename, delim_whitespace=True, names=['fromNode','toNode'])
print(pipes.head())
fromX = nodes.X.loc[pipes.fromNode]
fromY = nodes.Y.loc[pipes.fromNode]
toX   = nodes.X.loc[pipes.toNode]
toY   = nodes.Y.loc[pipes.toNode]
f,ax = plt.subplots(1,2, figsize=(15,10))
ax[0].scatter(fromX, fromY, marker='.')
ax[0].set_aspect('equal')
ax[1].scatter(toX,   toY,   marker='.')
ax[1].set_aspect('equal')
```
